```python
indcases = pd.read_excel(r"C:\Users\jaskeerat singh\Desktop\Covid cases in India.xlsx")
```

```python
indcases.drop("S. No.", axis = 1 , inplace = True)
indcases
```

```python
indcases["total cases"] = indcases["Total Confirmed cases (Indian National)"] + indcases["Total Confirmed cases ( Foreign National )"]
indcases
```

```python
indcases["total cases"].sum()
```

```python
indcases["active cases"] = indcases["total cases"] - indcases["Cured"] - indcases["Death"]
indcases
```

```python
indcases.style.background_gradient(cmap = "Blues")
```

```python
indcases.sort_values(by = "active cases", ascending = False).style.background_gradient(cmap = "Reds")
```

```python
plt.bar(indcases["Name of State / UT"],indcases["active cases"],label= "Number of cases", color = "#fcb503")
plt.title("Active cases in each state")
plt.xlabel("States/UT")
plt.xlabel("Active cases")
plt.legend()
plt.grid(True)
plt.xticks(rotation= 270)
plt.show()
```

```python
indcases.head()
```

```python
plt.scatter(indcases["Name of State / UT"],indcases["total cases"], label = "Surge of Cases", color = "#fc0303",edgecolors = "Black")
plt.plot(indcases["Name of State / UT"],indcases["total cases"])
plt.title("Scatter plot")
plt.xlabel("States")
plt.ylabel("Number of cases")
plt.legend()
plt.grid(True)
plt.xticks(rotation= 270)
plt.show()
```

```python
worlddata = pd.read_csv(r"C:\Users\jaskeerat singh\Desktop\worldcovid.csv")
worlddata
```

```python
worlddata = worlddata.drop("SNo",axis = 1,inplace = True)
worlddata
```

```python
wdata = worlddata.groupby("Country/Region").sum()
wdata
```

```python
wdata = wdata.sort_values(by = "Confirmed",ascending = False)
wdata
```

```python
ywdata = wdata["Confirmed"].head(20).to_frame()
a = ywdata.drop("Mainland China")
a
```

```python
plt.plot(a,color = "yellow" ,linewidth = 5,label = "Plot - line")
plt.title("Plot of countries and covid patients")
plt.xlabel("Countries")
plt.ylabel("Number of corona cases")
plt.legend()
plt.grid(True)
plt.xticks(rotation= "vertical")
plt.show()
```

```python
plt.bar(a.index,a["Confirmed"],color = "red" ,label = "bargraph")
plt.title("Plot of countries and covid patients")
plt.xlabel("Countries")
plt.ylabel("Number of corona cases")
plt.legend()
plt.grid(True)
plt.xticks(rotation= "vertical")
plt.show()
```
